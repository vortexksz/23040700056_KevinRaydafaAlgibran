```python
import hashlib
import os
from scapy.all import rdpcap

# daftar file
pcap_files = [
    "PCAP01_23040700056.pcap",
    "PCAP02_23040700056.pcap",
    "PCAP03_23040700056.pcap",
    "PCAP04_23040700056.pcap",
    "PCAP05_23040700056.pcap"
]

for file in pcap_files:

    # menghitung SHA-256
    sha256 = hashlib.sha256()

    with open(file, "rb") as f:
        while chunk := f.read(4096):
            sha256.update(chunk)

    hash_value = sha256.hexdigest()

    # ukuran file
    file_size = os.path.getsize(file)

    # jumlah paket
    packets = rdpcap(file)
    packet_count = len(packets)

    print("=" * 60)
    print("Nama File   :", file)
    print("Jumlah Paket:", packet_count)
    print("Ukuran File :", file_size, "bytes")
    print("SHA-256     :", hash_value)
```

```python
import hashlib
import datetime
import os
from scapy.all import rdpcap
```

```python
class Block:

    def __init__(
        self,
        index,
        timestamp,
        evidence_file,
        packet_count,
        evidence_hash,
        previous_hash
    ):

        self.index = index
        self.timestamp = timestamp
        self.evidence_file = evidence_file
        self.packet_count = packet_count
        self.evidence_hash = evidence_hash
        self.previous_hash = previous_hash

        self.block_hash = self.calculate_hash()

    def calculate_hash(self):

        data = (
            str(self.index)
            + str(self.timestamp)
            + str(self.evidence_file)
            + str(self.packet_count)
            + str(self.evidence_hash)
            + str(self.previous_hash)
        )

        return hashlib.sha256(data.encode()).hexdigest()
```

```python
def calculate_file_hash(filename):

    sha256 = hashlib.sha256()

    with open(filename, "rb") as f:
        sha256.update(f.read())

    return sha256.hexdigest()
```

```python
genesis_block = Block(
    index=0,
    timestamp=datetime.datetime.now(),
    evidence_file="GENESIS",
    packet_count=0,
    evidence_hash="0",
    previous_hash="0"
)
```

```python
blockchain = [genesis_block]
```

```python
pcap_files = [
    "PCAP01_23040700056.pcap",
    "PCAP02_23040700056.pcap",
    "PCAP03_23040700056.pcap",
    "PCAP04_23040700056.pcap",
    "PCAP05_23040700056.pcap"
]
```

```python
for file in pcap_files:

    evidence_hash = calculate_file_hash(file)

    packet_count = len(rdpcap(file))

    previous_hash = blockchain[-1].block_hash

    block = Block(
        index=len(blockchain),
        timestamp=datetime.datetime.now(),
        evidence_file=file,
        packet_count=packet_count,
        evidence_hash=evidence_hash,
        previous_hash=previous_hash
    )

    blockchain.append(block)
```

```python
for block in blockchain:

    print("=" * 80)

    print("Index         :", block.index)
    print("Timestamp     :", block.timestamp)
    print("Evidence File :", block.evidence_file)
    print("Packet Count  :", block.packet_count)
    print("Evidence Hash :", block.evidence_hash)
    print("Previous Hash :", block.previous_hash)
    print("Block Hash    :", block.block_hash)
```

```python
def validate_blockchain(blockchain):

    # mulai dari block ke-1
    # karena Genesis Block tidak punya block sebelumnya

    for i in range(1, len(blockchain)):

        current_block = blockchain[i]
        previous_block = blockchain[i - 1]

        # ---------------------------
        # 1. Validasi Previous Hash
        # ---------------------------

        if current_block.previous_hash != previous_block.block_hash:

            print(f"ERROR pada Block {current_block.index}")
            print("Previous Hash tidak sesuai")

            return False

        # -----------------------
        # 2. Validasi Block Hash
        # -----------------------

        recalculated_hash = current_block.calculate_hash()

        if current_block.block_hash != recalculated_hash:

            print(f"ERROR pada Block {current_block.index}")
            print("Block Hash tidak sesuai")

            return False

    # ---------------------------------
    # 3. Integritas blockchain terjaga
    # ---------------------------------

    return True
```

```python
if validate_blockchain(blockchain):
    print("Blockchain Validation : VALID")
else:
    print("Blockchain Validation : INVALID")
```
